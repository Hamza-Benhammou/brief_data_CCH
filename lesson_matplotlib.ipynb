```python
import numpy as np
import matplotlib.pyplot as plt
```

#### 1) Pyplot

Très simple à utiliser. Elle nous vient du module 'matplotlib.pyplot'

On va entrer des données qu'on veut afficher sur un axe X (abscisse) et des données qu'on veut afficher sur un axe Y (ordonnés)


- Ici, nous avons 2 tableaux numpy :

```python
# un tableau X qui comprend 10 points allant de 0 à 2
x = np.linspace(0, 2, 10)
# et un tableau Y qyui est le carré de X
y = x**2
print(x)
```

```python
# si je veux placer Y en fonction de X, je vais tracer X sur les basisse et Y sur les ordonnés :
plt.plot(x, y)

#  et j'obtiens un graĥique
```

`/!\ Si on devait utiliser autre chose que JUPYTER NOTEBOOK, le graphique ne s'affichera pas comme ça. Il faudra écrire :`
plt.show()

```python
plt.plot(x, y)
plt.show() # permets d'afficher le graphique, tout simplement
```

#### Voici la chose la plus importante à comprendre avec Matplotlib : 
#### LE CYCLE DE VIE D'UNE FIGURE

Jusqu'ici quand on a créer notre figure, on a directement écrit 'plt.plot()'.

Certes c'est rapide, simple et efficace MAIS EN PRINCIPE, quand on veut créer une figure, on ne commence pas par écrire 'plt.plot()' :\
On commence par écrire 'plt.figure()' et ça, ça nous crée une nouvelle figure TOUTE VIDE !

Une figure dans Matplotlib, ça représente simplement une feuille de travail = `La fenêtre de travail` sur lequelle on va travailler :

```python
plt.figure()
# Donc si on compile ce code, rien ne s'affiche (=NORMAL) parce qu'on une figure sans aucun axe
```

```python
plt.figure()
# Une fois qu'on a créer une figure, on va pouvoir créer
plt.plot(x, y)
# ce qui nous donne notre figure avec notre courbe dessus
```

Maintenant la raison pour laquelle on a explicitement écrit notre figure c'est parce que à l'intérieur des parenthèses on va par exemple pouvoir choisir la taille (À RETENIR) :

```python
plt.figure(figsize=(12, 8)) # on rentre entre () les dimensions de notre figure, /!\ les dimensions sont en 'inch' ! Pas en 'cm'
# Faire apparaître 2 courbes :
plt.plot(x, y, label="quadratique")
plt.plot(x, x**3, label="cubique")
# Ajouter un titre :
plt.title("Mon premier graphique")
# Donner des noms à nos axes :
plt.xlabel("absisse")
plt.ylabel("ordonnés")
# Ajouter une légende (on rajoute ci-haut pour chaque courbe label="quadratique" et label="cubique")
plt.legend()
# Et on fini évidemment par :
plt.show()
# Option pour sauvegarder notre figure, une fois qu'elle est terminée :
plt.savefig("figure.png") # qu'on va retrouver dans notre répertoire de travail (ctrl+B)
```

#### 2) Subplot

Si on veut afficher plusieurs graphiques sur une seule et même figure, c'est possible !\
Et pour cela, on utilisera la fonction 'subplot()'

Elle nous permet de générer une grille de différents graphiques sur notre figure.\
- A l'intérieur de ce 'subplot()' on va écrire le nombre de lignes et de colonnes qu'on va avoir dans notre grille
- Ensuite on va écrire sur quel graphique à l'intérieur de cette grille on a envie de travailler

Par exemple :

```python
# On veut créer 2 graphiques au sein d'une figure :
# On veut créer deux graphiques sur une même colonne (visuellement, il s'afficheront : un au dessus, et l'autre en dessous du premier)
# Premier graphique :
plt.subplot(2, 1, 1) # (2 lignes, 1 colonne, graphique sur lequel on travail)
plt.plot(x, y, c="red")
# Deuxième graphique :
plt.subplot(2, 1, 2) # (2 lignes, 1 colonne, graphique sur lequel on travail)
plt.plot(x, y, c="blue")
plt.show()
```
