```python
import pandas as pd
df_anim = pd.read_csv("data_animals.csv")
df_veg = pd.read_csv("data_vegetals.csv")
```

1. Nettoyer les titres de colonnes :\
Supprimer les espaces au début et à la fin des titres\
Remplacer les espaces par des underscores (ceux se situant entre les mots)\
Tout mettre en minuscule


```python
df_anim.columns = [title.lower().strip().replace(" ", "_") for title in df_anim.columns]
df_veg.columns = [title.lower().strip().replace(" ", "_") for title in df_veg.columns]
print(df_veg, df_anim)
```
