```python
list_reg = ['eastern_africa', 'middle_africa', 'northern_africa', 'southern_africa', 'western_africa', 
          'northern_america', 'central_america', 'caribbean', 'south_america', 'central_asia', 
          'eastern_asia', 'southern_asia', 'south_eastern_asia', 'western_asia', 'eastern_europe', 
          'northern_europe', 'southern_europe', 'western_europe', 'australia_new_zealand', 'melanesia', 
          'micronesia', 'polynesia']
```

```python
def transform_data(filename=None):
    df = pd.read_csv("Regions/"+filename+".csv", usecols=["Area Code (M49)"])
    df["zone"] = filename
    return df

def aggregate_df(list_filename=list_reg):
    list_df = []
    for zone in list_filename:
        temp = transform_data(zone)
        list_df.append(temp)

    df = pd.concat(list_df)
    return df

zone = aggregate_df()
```

`ma partie :`

`je modifie le titre de la colonne 'Area Code (M49)' en 'area_code_m49'`

```python
zone.columns = [title.lower().strip().replace(" ", "_").replace("(", "").replace(")", "").replace("/", "_") for title in zone.columns]
display(zone)
```

`Puis je modifie la colonne 'area_code_m49' par 'zone'`

```python
zone = zone.rename(columns={"area_code_m49": "country_code"})
display(zone)
```

`Enfin je merge`

```python
df = pd.merge(zone, df)
display(df)
```
