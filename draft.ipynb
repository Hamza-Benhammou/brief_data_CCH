```python
# Question 16
# Nous allons ajouter une nouvelle colonne à notre jeu de données : la zone géographique. 
# Pour cela rendez-vous sur le site de la FAO, votre objectif est de télécharger des nouveaux datasets
# pour ajouter les zones suivantes à votre jeu de données : 
# 'eastern_africa', 'middle_africa', 'northern_africa', 'southern_africa', 'western_africa', 'northern_america', 
# 'central_america', 'caribbean', 'south_america', 'central_asia', 'eastern_asia', 'southern_asia',
# 'south_eastern_asia', 'western_asia', 'eastern_europe', 'northern_europe', 'southern_europe', 
# 'western_europe', 'australia_new_zealand', 'melanesia', 'micronesia', 'polynesia

regions = ['eastern_africa', 'middle_africa', 'northern_africa', 'southern_africa', 'western_africa', 'northern_america', 'central_america', 'caribbean', 'south_america', 'central_asia', 'eastern_asia', 'southern_asia', 'south_eastern_asia', 'western_asia', 'eastern_europe', 'northern_europe', 'southern_europe', 'western_europe', 'australia_new_zealand', 'melanesia', 'micronesia', 'polynesia'
]
def new_data(filename=None):
    df = pd.read_csv('regions/'+filename+".csv", usecols=["Area Code (M49)"])
    df['zone_geographique'] = filename
    return df

def add_all_files(list_filename = regions):
    list_df = []
    for zone_geogrpahique in list_filename:
        temp = new_data(zone_geogrpahique)
        list_df.append(temp)
    df = pd.concat(list_df)
    df.columns = ['country_code', 'zone_geographique']
    
    return df

df_global = add_all_files()
df = pd.merge(df, df_global)


display(df)
```

```python
# Question 24
# Calculez la disponibilité de nourriture totale par pays et par année, en kcal et kg de protéines. Attention aux unités de mesure !




# disponibilité totale en kcal par habitant
df['disponibilite_kcal_par_habitant'] = df['food_supply_(kcal/capita/day)'] * 365  # Convertir de kcal/jour à kcal/an
df['disponibilite_kcal_par_habitant'] = df['disponibilite_kcal_par_habitant'] * 1e3  # Convertir de kcal à kcal par 1000 habitants

# disponibilité totale de protéines par habitant
df['disponibilite_proteines_kg_par_habitant'] = df['protein_supply_quantity_(g/capita/day)'] * 365  # Convertir de g/jour à g/an
df['disponibilite_proteines_kg_par_habitant'] = df['disponibilite_proteines_kg_par_habitant'] / 1e3  # Convertir de g à kg

# par pays et par année 
disponibilite_totale = df.groupby(['country', 'year'])[['disponibilite_kcal_par_habitant', 'disponibilite_proteines_kg_par_habitant']].sum().reset_index()

display(disponibilite_totale)
```

```python
# Question 25 
# En vous basant sur le résultat de la question précédente, calculez le ratio énergie/poids de chaque produit et pays.
# Vous devriez vous apercevoir qu’étonnement, ces informations varient en fonction du pays. Pour pallier ce problème, 
# calculez la moyenne de ce ratio pour chaque aliment.  Attention à bien gérer les valeurs égales à 0. 
# Vérifiez la cohérence de votre calcul en comparant le résultat avec l’apport calorique d’un œuf.


# Calculer le ratio énergie/poids pour chaque produit et pays
df['ratio_energie_poids'] = df['food_supply_(kcal/capita/day)'] / df['food_supply_quantity_(kg/capita/yr)']

# Remplacer les valeurs infinies ou NaN par 0
df['ratio_energie_poids'].replace([np.inf, -np.inf, np.nan], 0, inplace=True)

# Calculer la moyenne du ratio énergie/poids pour chaque aliment
moyenne_ratio_par_aliment = df.groupby('item')['ratio_energie_poids'].mean().reset_index()

# Afficher le résultat
display(moyenne_ratio_par_aliment)
```

```python
# Question 26
# À l’instar de la question précédente, calculez le pourcentage de protéine de chaque aliment. 
# Vérifiez votre résultat en le comparant avec l’apport en protéines d’un œuf.



# pourcentage de protéines pour chaque aliment
df['pourcentage_proteines'] = (df['protein_supply_quantity_(g/capita/day)'] / df['food_supply_quantity_(kg/capita/yr)']) * 100

# Remplacer les valeurs infinies ou NaN par 0
df['pourcentage_proteines'].replace([np.inf, -np.inf, np.nan], 0, inplace=True)

pourcentage_proteines_par_aliment = df.groupby('item')['pourcentage_proteines'].mean().reset_index()

print(pourcentage_proteines_par_aliment)
```
