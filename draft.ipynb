Si certaines des colonnes ne nous intéresse pas forcément on va pouvoir les éliminer :

(imaginons que les colonnes 'Domain Code' et 'Domain' ne nous intéresse pas)

```python
data = pd.read_csv("data_vegetals.csv")
data = data.drop(["Domain", "Domain Code"], axis=1)
data.head()
```

Produire les statistiques de bases pour nos colonnes :

```python
data.describe()
```

Le 'count' (ci-haut) indique le nombre de ligne. Ici dans chaque colonnes, on voit que le nombre de colonnes est identique. 86226 dans notre cas.

Si dans une des colonnes (par exemple: prenons la colonne 'Value') le nombre de lignes était inférieur à 86226 ici.\
Ceci nous indique qu'il nous manque des données concernant 'Value'. 

Et quand il nous manque des données, c'est donc difficile de faire de l'analyse de données.

Dans ces situations, on 2 options :

OPTION 1 :\
Utiliser la fonction fillna() pour remplacer les données manquantes par une valeur par défaut.\
Par exemple, on pourrait dire que quand il nous manque une donnée, on pourrait tout simplement remplacer ça par une moyenne

--> *Ça peut être une bonne option MAIS ATTENTION il faut garder à l'esprit que ça va corrompre notre dataset du coup on peut opter pour l'OPTION 2*

```python
data.fillna(data["Value"].mean())
```

OPTION 2 :
Éliminer les lignes dans lesquelles ils nous manquent des données\

--> *Parfois il vaut mieux perdre un peu de données plutôt que de corrompre la réalité des choses*

```python
data.dropna(axis=0)
```

Maintenant, mettons qu'on veux savoir combien il y a de produits dans les différents types d'éléments (= disponibilité alimentaire en kg, disponibilité alimentaire en kcal, disponibilité alimentaire en protéines, disponibilité alimentaire en matière grasse).

On va donc utiliser la fonction 'value_counts()'

```python
data["Element"].value_counts()
```

Maintenant avec ces données, nous allons créer notre premier graphique (simple)

```python
data["Element"].value_counts().plot.bar()
```
