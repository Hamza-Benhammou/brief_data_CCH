```python
# Question 16
# Nous allons ajouter une nouvelle colonne Ã  notre jeu de donnÃ©es : la zone gÃ©ographique. 
# Pour cela rendez-vous sur le site de la FAO, votre objectif est de tÃ©lÃ©charger des nouveaux datasets
# pour ajouter les zones suivantes Ã  votre jeu de donnÃ©es : 
# 'eastern_africa', 'middle_africa', 'northern_africa', 'southern_africa', 'western_africa', 'northern_america', 
# 'central_america', 'caribbean', 'south_america', 'central_asia', 'eastern_asia', 'southern_asia',
# 'south_eastern_asia', 'western_asia', 'eastern_europe', 'northern_europe', 'southern_europe', 
# 'western_europe', 'australia_new_zealand', 'melanesia', 'micronesia', 'polynesia

regions = ['eastern_africa', 'middle_africa', 'northern_africa', 'southern_africa', 'western_africa', 'northern_america', 'central_america', 'caribbean', 'south_america', 'central_asia', 'eastern_asia', 'southern_asia', 'south_eastern_asia', 'western_asia', 'eastern_europe', 'northern_europe', 'southern_europe', 'western_europe', 'australia_new_zealand', 'melanesia', 'micronesia', 'polynesia'
]
def new_data(filename=None):
    df = pd.read_csv('regions/'+filename+".csv", usecols=["Area Code (M49)"])
    df['zone_geographique'] = filename
    return df

def add_all_files(list_filename = regions):
    list_df = []
    for zone_geogrpahique in list_filename:
        temp = new_data(zone_geogrpahique)
        list_df.append(temp)
    df = pd.concat(list_df)
    df.columns = ['country_code', 'zone_geographique']
    
    return df

df_global = add_all_files()
df = pd.merge(df, df_global)


display(df)
```

```python
# Question 24
# Calculez la disponibilitÃ© de nourriture totale par pays et par annÃ©e, en kcal et kg de protÃ©ines. Attention aux unitÃ©s de mesure !




# disponibilitÃ© totale en kcal par habitant
df['disponibilite_kcal_par_habitant'] = df['food_supply_(kcal/capita/day)'] * 365  # Convertir de kcal/jour Ã  kcal/an
df['disponibilite_kcal_par_habitant'] = df['disponibilite_kcal_par_habitant'] * 1e3  # Convertir de kcal Ã  kcal par 1000 habitants

# disponibilitÃ© totale de protÃ©ines par habitant
df['disponibilite_proteines_kg_par_habitant'] = df['protein_supply_quantity_(g/capita/day)'] * 365  # Convertir de g/jour Ã  g/an
df['disponibilite_proteines_kg_par_habitant'] = df['disponibilite_proteines_kg_par_habitant'] / 1e3  # Convertir de g Ã  kg

# par pays et par annÃ©e 
disponibilite_totale = df.groupby(['country', 'year'])[['disponibilite_kcal_par_habitant', 'disponibilite_proteines_kg_par_habitant']].sum().reset_index()

display(disponibilite_totale)
```

```python
# Question 25 
# En vous basant sur le rÃ©sultat de la question prÃ©cÃ©dente, calculez le ratio Ã©nergie/poids de chaque produit et pays.
# Vous devriez vous apercevoir quâ€™Ã©tonnement, ces informations varient en fonction du pays. Pour pallier ce problÃ¨me, 
# calculez la moyenne de ce ratio pour chaque aliment.  Attention Ã  bien gÃ©rer les valeurs Ã©gales Ã  0. 
# VÃ©rifiez la cohÃ©rence de votre calcul en comparant le rÃ©sultat avec lâ€™apport calorique dâ€™un Å“uf.


# Calculer le ratio Ã©nergie/poids pour chaque produit et pays
df['ratio_energie_poids'] = df['food_supply_(kcal/capita/day)'] / df['food_supply_quantity_(kg/capita/yr)']

# Remplacer les valeurs infinies ou NaN par 0
df['ratio_energie_poids'].replace([np.inf, -np.inf, np.nan], 0, inplace=True)

# Calculer la moyenne du ratio Ã©nergie/poids pour chaque aliment
moyenne_ratio_par_aliment = df.groupby('item')['ratio_energie_poids'].mean().reset_index()

# Afficher le rÃ©sultat
display(moyenne_ratio_par_aliment)
```

```python
# Question 26
# Ã€ lâ€™instar de la question prÃ©cÃ©dente, calculez le pourcentage de protÃ©ine de chaque aliment. 
# VÃ©rifiez votre rÃ©sultat en le comparant avec lâ€™apport en protÃ©ines dâ€™un Å“uf.



# pourcentage de protÃ©ines pour chaque aliment
df['pourcentage_proteines'] = (df['protein_supply_quantity_(g/capita/day)'] / df['food_supply_quantity_(kg/capita/yr)']) * 100

# Remplacer les valeurs infinies ou NaN par 0
df['pourcentage_proteines'].replace([np.inf, -np.inf, np.nan], 0, inplace=True)

pourcentage_proteines_par_aliment = df.groupby('item')['pourcentage_proteines'].mean().reset_index()

print(pourcentage_proteines_par_aliment)
```
