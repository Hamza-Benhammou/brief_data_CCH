On enregistre notre DataFrame sous une variable :

```python
data= pd.read_csv("data_vegetals.csv")
```

```python
data.shape

# Renvoi un tuple dans lequel est affiché le nombre de lignes = 86226, et le nombre de colonnes = 14
```

Voir les différentes colonnes :

Afficher le début de votre DataFrame : 

```python
data.head()

# Entre () on peut indiquer le nombre de lignes qui nous intéresse (indices fonctionne à peu près comme pour les listes)
```

Nettoyer les données :

```python
data.columns = [title.lower().strip().replace(" ", "_").replace("(", "").replace(")", "") for title in data.columns]
display(data)
```

Si certaines des colonnes ne nous intéresse pas forcément on va pouvoir les éliminer :

(imaginons que les colonnes 'Domain Code' et 'Domain' ne nous intéresse pas)

```python
data = data.drop(["domain", "domain_code"], axis=1)
data.head()
```

Produire les statistiques de bases pour nos colonnes :

```python
data.describe()
```

Le 'count' (ci-haut) indique le nombre de ligne. Ici dans chaque colonnes, on voit que le nombre de colonnes est identique. 86226 dans notre cas.

Si dans une des colonnes (par exemple: prenons la colonne 'Value') le nombre de lignes était inférieur à 86226 ici.\
Ceci nous indique qu'il nous manque des données concernant 'Value'. 

Et quand il nous manque des données, c'est donc difficile de faire de l'analyse de données.

Dans ces situations, on 2 options :

OPTION 1 :\
Utiliser la fonction fillna() pour remplacer les données manquantes par une valeur par défaut.\
Par exemple, on pourrait dire que quand il nous manque une donnée, on pourrait tout simplement remplacer ça par une moyenne

--> *Ça peut être une bonne option MAIS ATTENTION il faut garder à l'esprit que ça va corrompre notre dataset du coup on peut opter pour l'OPTION 2*

```python
data.fillna(data["value"].mean())

# mean() = la moyenne
```

OPTION 2 :
Éliminer les lignes dans lesquelles ils nous manquent des données\

--> *Parfois il vaut mieux perdre un peu de données plutôt que de corrompre la réalité des choses*

```python
data.dropna(axis=0)
```

Maintenant, mettons qu'on veux savoir combien il y a de produits dans les différents types d'éléments (= disponibilité alimentaire en kg, disponibilité alimentaire en kcal, disponibilité alimentaire en protéines, disponibilité alimentaire en matière grasse).

On va donc utiliser la fonction 'value_counts()'

```python
data["element"].value_counts()
```

Maintenant avec ces données, nous allons créer notre premier graphique (simple)

```python
data["element"].value_counts().plot.bar()
```

### L'envers du décor

Dans Pandas, il existe deux structures de données : le DataFrame et la Série

Simplement, une série c'est juste une colonne dans un DataFrame
Et quand on assemble plusieurs Séries les unes à côté des autres ça nous donne un nouveau DataFrame

Une Série plus précisémment c'est un tableau NumPy à une dimension à côté duquel on a rajouté une colonne 
(voir 10min 42s de la vidéo : PANDAS PYTHON Français - Introduction + Analyse du Titanic (17/30) - Machine Learnia) --> suite cahier (14/11/2023)

```python
data = data.set_index("element_code")
data["area"]
```

On peut dire qu'un DataFrame, c'est un peu comme un DICTIONNAIRE  qui contient des Séries. Il faut imaginer :

dict["clé"] = valeur


Différentes opérations d'indexing et de slicing qu'on peut retrouver dans Pandas :

`data["column_name"]` *= Série (ndarray)*

`data["column_name"][0:10]` *(indexing)*

`data["column_name"] < 18` *(max) = va retourner des True et des False* 

`data[data["column_name"] < 18]` *(boolean indexing) = va nous retourner tous les éléments inférieur à 18*

`data[data["column1_name", "column2_name"]]` = *DataFrame*



#### Les fonctions 'iloc' et 'loc' :

RAPPEL : Si vous êtes perdu, rappelez-vous que :  `data["column_name"][0:10]`\
`data["column_name"]`[0:10] => ce qui se trouve à **gauche** ça va être une **Série** \
data["column_name"]`[0:10]` => et à **droite** ça va être l'**indexing**

`data["column_name"][0:10]` Si on exécute ce code, 
**on obtient les 10 premières valeurs de notre Série "column_name"**

```python
data.iloc[0:2, 0:2] #--> localisation par index

# iloc veut dire (en anglais) index localisation
```
